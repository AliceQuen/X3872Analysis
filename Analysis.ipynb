# Step A1: Generating Trees From Nutples
This step generates *Source Tree* for cut optimization, TMVA application and side band subtraction distribution.

The Tree looks like this
- $\psi(2S)2\mu 2\pi$ mass
- $J/\psi_2$ mass for 2D fitting
- $X(6900) 4\mu 2\pi$ mass with track fitting
- $X(6900)$ fit vertex probablity
- $X(6900) 4\mu 2\pi$ mass with Lorentz vector
- other variables

The four variables above is used in the TMVA and cut optimization application, please do specific them. There is a default defination of them.
